# Functions 

```python
def makeDataset(url):
    with open("results/dataset.csv", "w") as f:
        f.write(session.get(url).text)

    with open("results/dataset.csv", "r") as f:
        dataset = csv.reader(f)
        niftyData = []
        stockData = []

        for idx, row in enumerate(dataset):
            if 8 <= idx <= 67:
                niftyData.append(row)
            if 120 <= idx:
                stockData.append(row)
    os.remove("results/dataset.csv")
    return pd.DataFrame(niftyData), pd.DataFrame(stockData)
```

```python
def getTodayData() -> object:
    url = "https://www.nseindia.com/api/merged-daily-reports?key=favCapital"
    webData = session.get(url=url, headers=head)
    return makeDataset(webData.json()[1]['link'])
```
